```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Параметр масштаба (shape parameter) распределения Рэлея
scale_parameter = 2.0

# Генерация данных для распределения Рэлея
data = np.random.rayleigh(scale=scale_parameter, size=1000)

# Построение гистограммы
plt.hist(data, bins=30, density=True, alpha=0.7, color='b')

# Построение теоретической плотности распределения
x = np.linspace(0, 10, 100)
pdf = (x / scale_parameter**2) * np.exp(-x**2 / (2 * scale_parameter**2))
plt.plot(x, pdf, 'r-', label='Theory')

# Настройка графика
plt.title('Rayleigh Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Отображение графика
plt.show()
```
