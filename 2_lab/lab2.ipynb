Импорт библиотек numpy, scipy, matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

# Построение распределений

```python
# Параметр масштаба (shape parameter) распределения Рэлея
scale_parameter = 2.0

# Генерация данных для распределения Рэлея
data = np.random.rayleigh(scale=scale_parameter, size=1000)

# Построение гистограммы
plt.hist(data, bins=30, density=True, alpha=0.7, color='b')

# Построение теоретической плотности распределения
x = np.linspace(0, 10, 100)
pdf = (x / scale_parameter**2) * np.exp(-x**2 / (2 * scale_parameter**2))
plt.plot(x, pdf, 'r-', label='Theory')

# Настройка графика
plt.title('Rayleigh Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Отображение графика
plt.show()
```

# Построение графика на по функции

```python
# Генерация данных для функции (x^2 + y^2 - 1)^3 - x^2 y^3 = 0
def generate_data():
    x = np.linspace(-2, 2, 400)
    y = np.linspace(-2, 2, 400)

    X, Y = np.meshgrid(x, y)
    Z = (X**2 + Y**2 - 1)**3 - X**2 * Y**3

    return X, Y, Z

# Построение графика
def plot_graph(X, Y, Z):
    plt.contour(X, Y, Z, levels=[0], colors='r')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('(x^2 + y^2 - 1)^3 - x^2 y^3 = 0')

    # Добавление сетки на задний план
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.show()

# Генерация данных
X, Y, Z = generate_data()

# Построение графика
plot_graph(X, Y, Z)
```

# Анализ зависимостей

```python
import csv

data = []
with open('dataset_home.txt', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        data.append(row)
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
for row in data:
    age.append(int(row[0]))
    sex.append(row[1])
    bmi.append(float(row[2]))
    children.append(int(row[3]))
    smoker.append(row[4])
    region.append(row[5])
    charges.append(float(row[6]))
```

```python
# Scatter plot для Возраста и Расходов
plt.scatter(age, charges, alpha=0.5)
plt.title('Scatter Plot: Возраст vs Расходы')
plt.xlabel('Возраст')
plt.ylabel('Расходы')
plt.show()
```

Исходя из результатов, показанных на графике, можно заметить корреляцию между тем, как возраст влияет на расходы. Можно предположить, что с возрастом расходы увеличиваются.


```python
# Scatter plot для Возраста и количества детей
plt.scatter(age, children, alpha=0.5)
plt.title('Scatter Plot: Возраст vs Дети')
plt.xlabel('Возраст')
plt.ylabel('Дети')
plt.show()
```

Исходя из результатов, показанных на графике, можно заметить что в наборе данных типичные значения для показателей возраста и количества детей

```python
# Линейный график для Расходов
plt.plot(charges)
plt.title('Line Plot: Расходы')
plt.xlabel('Index')
plt.ylabel('Расходы')
plt.show()
```

Линейный график показывает, изменение расходов

```python
# Гистограмма для Возраста
plt.hist(age, bins=20, edgecolor='black')
plt.title('Histogram: Возраст')
plt.xlabel('Возраст')
plt.ylabel('Frequency')
plt.show()
```

Гистограмма возраста демонстрирует распределение в наборе данных

```python
# Bar chart для Smoker
unique_smokers, counts = np.unique(smoker, return_counts=True)
plt.bar(unique_smokers, counts, color=['blue', 'red'])
plt.title('Bar Chart: Smoker')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```

Данный график показывает, сколько в данных курящих/не курящих людей

```python
# Визуализация корреляций с использованием тепловой карты
correlation_matrix = np.corrcoef([age, bmi, charges])
plt.figure(figsize=(8, 6))
heatmap = plt.matshow(correlation_matrix, cmap="coolwarm")
plt.colorbar(heatmap, label="Correlation Coefficient")
plt.title("Correlation Heatmap")
plt.show()
# Рассчет корреляции между Age и Charges
correlation, _ = pearsonr(age, charges)
print(f'Correlation between Age and Charges: {correlation}')
# Рассчет корреляции между BMI и Charges
correlation, _ = pearsonr(bmi, charges)
print(f'Correlation between BMI and Charges: {correlation}')
```

Тепловая карта показывает взаимосвязь между возрастом, индексом массы тела и расходов. Темные ячейки показывают более сильную кореляцию.

```python
# Гистограмма для BMI
plt.hist(bmi, bins=20, edgecolor='black')
plt.title('Histogram: BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

Гистограмма индекса массы тела демонстрирует распределение в наборе данных.
