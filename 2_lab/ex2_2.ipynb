```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация данных для функции (x^2 + y^2 - 1)^3 - x^2 y^3 = 0
def generate_data():
    x = np.linspace(-2, 2, 400)
    y = np.linspace(-2, 2, 400)

    X, Y = np.meshgrid(x, y)
    Z = (X**2 + Y**2 - 1)**3 - X**2 * Y**3

    return X, Y, Z

# Построение графика
def plot_graph(X, Y, Z):
    plt.contour(X, Y, Z, levels=[0], colors='r')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('(x^2 + y^2 - 1)^3 - x^2 y^3 = 0')

    # Добавление сетки на задний план
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.show()

# Генерация данных
X, Y, Z = generate_data()

# Построение графика
plot_graph(X, Y, Z)
```
