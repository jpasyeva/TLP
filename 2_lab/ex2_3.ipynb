```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Загрузка данных из внешнего источника
url = "https://lhd8ts4trw.a.trbcdn.net/cdn/yusuciqe45/00_data_engineer/dataset_home.txt"
data = np.genfromtxt(url, delimiter=',', dtype=None, encoding=None, names=True)

# Извлечение данных для каждой переменной
age = data["age"]
bmi = data["bmi"]
charges = data["charges"]

# Scatter plot для Age и Charges
plt.scatter(age, charges, alpha=0.5)
plt.title('Scatter Plot: Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

# Scatter plot для BMI и Charges
plt.scatter(bmi, charges, alpha=0.5)
plt.title('Scatter Plot: BMI vs Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()

# Линейный график для Charges
plt.plot(charges)
plt.title('Line Plot: Charges')
plt.xlabel('Index')
plt.ylabel('Charges')
plt.show()

# Гистограмма для Age
plt.hist(age, bins=20, edgecolor='black')
plt.title('Histogram: Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Гистограмма для BMI
plt.hist(bmi, bins=20, edgecolor='black')
plt.title('Histogram: BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()

# Bar chart для Smoker
unique_smokers, counts = np.unique(data["smoker"], return_counts=True)
plt.bar(unique_smokers, counts, color=['blue', 'red'])
plt.title('Bar Chart: Smoker')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()

# Визуализация корреляций с использованием тепловой карты
correlation_matrix = np.corrcoef([age, bmi, charges])
plt.figure(figsize=(8, 6))
heatmap = plt.matshow(correlation_matrix, cmap="coolwarm")
plt.colorbar(heatmap, label="Correlation Coefficient")
plt.title("Correlation Heatmap")
plt.show()
# Рассчет корреляции между Age и Charges
correlation, _ = pearsonr(age, charges)
print(f'Correlation between Age and Charges: {correlation}')

# Рассчет корреляции между BMI и Charges
correlation, _ = pearsonr(bmi, charges)
print(f'Correlation between BMI and Charges: {correlation}')
```
